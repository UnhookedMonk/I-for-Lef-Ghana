```python
#import libraries to be used in project
import snscrape.modules.twitter as sntwitter
import pandas as pd
```

```python
#set search parameters for snscrape
query = "i for lef ghana until:2022-11-19 since:2021-01-01"
limit = 5000

#initialize list to which tweets will be appended
tweets = []
```

```python
#instruct snscrape to pull tweets based on earlier parameters and append to earlier list
for tweet in sntwitter.TwitterSearchScraper(query).get_items():
    if len(tweets) == limit:
        break
    else:
        tweets.append([tweet.date, tweet.user.username, tweet.content,tweet.sourceLabel])
```

```python
#append tweets in list to dataframe using pandas
df = pd.DataFrame(tweets, columns=['Date', 'User', 'Tweet', 'Tweet Source'])
df
```

```python
df.to_csv('tweets2022_11_19.csv')
```
